# **RSA**

RSA hanya bisa men-generate key dengan panjangnya 1024 bits keatas, gak bisa kurang dari 1024

```python
keyPair = RSA.generate(1024)
pubKey = keyPair.publickey()
key = hex(pubKey.n).lstrip('0x')
print("Public key: ", key)
```
